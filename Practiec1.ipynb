## Question 1
##### given a list of numbers, print "valid slice"" only if the first two elements are equal to the last two element

## Question 2
##### Check palindrome using slicing in python

```python
s = input("Enter a string:  ")
s=s.lower()
if s==s[-1::-1]:
    print("Palindrome String")
else:
    print("Not a Palindrome")
```

## Question 3
##### Given a string, check whether the last character is a digit using slicing, if digit-> Ends with digit otherwise Ends with character / Ends with special symbol.

```python
s = input("Enter a string::  ")
if s[-1:].isdigit():
    print("Ends with digit")
elif s[-1:].isalpha():
    print("Ends with Character")
else:
    print("Ends with special symbol")
```

## Question 6
##### if string len >= 5 find middle character is vowel or not other wise print String is too shot

```python
s = input("Enter a String: ")
l1 = 'aeiou'
if len(s) > 4:
    if s[len(s)//2].isalpha():
        if s[len(s)//2].lower() in l1:
            print("Middle character is vowel")
        else:
            print("middle character is not vowel")
    elif s[len(s)//2].isdigit():
        print("Middle character is digit")
    else:
        print("Middle character is Special character")
else:
    print("String is too short")
```

### Question 7
##### starts with A and end with Z valid String else invalid String

```python
s = input("Enter a String:  ")
if s.startswith('A'):
    if s.endswith('Z'):
        print("Valid AZ string")
    else: 
        print("Starts with A but invalid end")
else:
    print("Invalid String")
```

### Question 9
###### If the list length is even
###### Compare first 2 elements and last 2 elements, Slice with same element otherwise slice with different elements.
###### Else print "Odd length list"

```python
li = [5, 6, 1, 2]

if len(li) % 2 != 0:
    print("Odd length list")
else:
    if li[:2] == li[-2:]:
        print("Slices are same")
    else:
        print("Slices are different")
```
