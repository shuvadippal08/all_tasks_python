## Single Level

```python
class Tenth:
    @classmethod
    def __init__(cls, name: str, roll: int, marks: int):
        cls.name = name
        cls.roll = roll
        cls.marks = marks
        print("Parent class constructor")

    @classmethod
    def _show_details(cls):
        print(f"{cls.name} {cls.roll} {cls.marks}")
        print("Parent class show details")

    @staticmethod
    def abcd() -> None:
        print("within parent class")


class Twelfth(Tenth):
    @classmethod
    def __init__(self, name: str, roll: int, marks: int, email: str, phone: int, addr: str, marks12: int) -> None:
        super().__init__(name, roll, marks)
        self.email = email
        self.phone = phone
        self.addr = addr
        self.marks12 = marks12
        print("Child class Constructor")

    @classmethod
    def _show_details(self):
        super()._show_details()
        print(f"{self.addr} {self.email} {self.phone} {self.marks12}")
        print("child class showdetails")


tw = Twelfth("Shuva", 404301, 85, "svhuadippal015@gmail.com",
             9883610081, "shila", 85)
tw._show_details()
tw.abcd()
```

## Multilevel


```python
class Bachelors(Twelfth):
    def __init__(self, name: str, roll: int, marks: int, email: str, phone: int, addr: str, marks12: int, clg: str) -> None:
        super().__init__(name, roll, marks, email, phone, addr, marks12)
        self.clg = clg

    def _show_details(self):
        super()._show_details()
        print(f"{self.clg}")


st = Bachelors("Shuva", 404301, 85, "shvuadippal015@gmail.com",
               9876543210, "shila", 85, "HIT")
st._show_details()
```

```python
class Tenth:
    def __init__(self, name: str, roll: int, marks: int):
        self.name = name
        self.roll = roll
        self.marks = marks

    def show_details(self):
        print(f"Name: {self.name}, Roll: {self.roll}, Marks: {self.marks}")


class Contact:
    def __init__(self, email: str, phone: int, addr: str):
        self.email = email
        self.phone = phone
        self.addr = addr

    def show_details(self):
        print(
            f"Address: {self.addr}, Email: {self.email}, Phone: {self.phone}")


class Twelfth(Tenth, Contact):
    def __init__(
        self,
        name: str,
        roll: int,
        marks: int,
        email: str,
        phone: int,
        addr: str
    ):
        super().__init__(name, roll, marks)
        Contact.__init__(self, email, phone, addr)

    def show_details(self):
        super().show_details()
        Contact.show_details(self)

s = Twelfth(
    "Shuvadip",
    21,
    90,
    "abc@gmail.com",
    9876543210,
    "Katwa"
)

s.show_details()

```
