```python
## question 1
import logging

logging.basicConfig(
    filename="vowelcount.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s - %(name)s",
    filemode="a",
    force=True,
)
s = input("enter a string: ")
def vowelcount(s:str)->int:
    """
    Counts the number of vowel in a given string
    Parameters:
        s (str) : Input String
    Returns:
        int: total number of vowels in the string
    """
    logging.info("Counting the number of vowels")
    v='aeiou'
    cnt=0
    for ch in s.lower():
        if ch in v:
            cnt+=1
    logging.info(f"Vowel count in String {s} is {cnt}")
    return cnt
print(vowelcount(s))
```

```python
import logging

logging.basicConfig(
    filename="countfreq.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s - %(name)s",
    filemode="a",
    force=True,
)

s = input("enter a string ")

def countfreq(s:str)->dict:
    """
    This is the function which count the frequency for each character in a string
    Parameters:
        s (str) : Input String
    Output:
        k (dict): dict(str, int)
    """
    k={}
    logging.info("Counting the frequency")
    for ch in s:
        k[ch] = s.count(ch)
    logging.info(f"The frequency for each character in string {s} is {k}")
    return k
print(countfreq(s))
```

```python
import logging

logging.basicConfig(
    filename="findsecondlargest.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s - %(name)s",
    filemode="a",
    force=True,
)


l = [1,2,3,5,6,8,9,8,6]

def findsecondlargest(l:list[int])->int:
    """
    This function is for finding the second largest element in an given array or list
    Parameters:
        l (list) --> takes list input
    Output:
        Give output as the integer :
        the second largest element
        list must be integer type
    """
    logging.info("Finding the secondlargest element in an array")
    first = 0
    second = 0
    for num in l:
        if num > first:
            second=first
            first=num
        elif num > second and num != first:
            second= num
    logging.info(f"Second largest number found: {second} in the list {l}")
    return second

findsecondlargest(l)
```

```python
import logging

logging.basicConfig(
    filename="findmissingfrom1ton.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s - %(name)s",
    filemode="a",
    force=True,
)

l=[1,2,3,4,5,7]
def findmissingfrom1ton(l:list[int])->int:
    """
    This is a function which finds the missing number from an array 1 to n
    Parameter:
        l :(list[int]) --> Input list
    Output:
        num: int--> the missing number
    """
    n = len(l)+1
    actsum=n*(n+1)//2
    s = sum(l)
    k=abs(actsum-s)

    logging.info(f"The number found: {k}")
    return k
findmissingfrom1ton(l)
```
