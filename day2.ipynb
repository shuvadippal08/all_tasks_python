```python
k = int(input("Enter a number: "))
num =  k
sum = 0
while k > 0:
    rem = k % 10
    sum = sum * 10 + rem
    k//=10

if sum > (int)(num):
    print("The reversed number is greater than original number")
elif sum == (int)(num):
    print("number is a palindrome number")
else:
    print("The reversed number is smaller than the original number")
```

```python
def evencount(num: int) -> int:
    if num == 0:    return 0
    digit = num % 10
    if digit % 2 == 0:  return 1 + evencount(num // 10)
    else:   return evencount(num // 10)

def oddcount(num: int) -> int:
    if num == 0:    return 0
    digit = num % 10
    if digit % 2 != 0:  return 1 + oddcount(num // 10)
    else:   return oddcount(num // 10)
k=int(input("Enter the number:   "))
print(oddcount(k),'    ',evencount(k))
```
