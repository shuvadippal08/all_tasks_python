##### Merge and concat

```python
data1 = {
    'name':['A','B','C','D'],
    'age':[23,46,60,30],
    'salary':[30000,10000000,80000000,50000]
}

data2 = {
    'name':['aA','Bb','cC','D'],
    'age':[21,48,55,30],
    'salary':[23000,15600000,75000000,50000]
}
```

```python
new_df1 = pd.DataFrame(data1)
new_df2 = pd.DataFrame(data2)
```

```python
new_df1.head()
```

```python
new_df2.head()
```

```python
pd.concat([new_df1,new_df2],ignore_index=True)
```

```python
pd.concat([new_df1,new_df2],ignore_index=True,axis=1)
```

```python
new_df = pd.concat([new_df1,new_df2],ignore_index=True)
new_df.head()
```

##### Merge

```python
pd.merge(new_df2,new_df2,on='age',how='right')
```
