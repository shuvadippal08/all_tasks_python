```python
no1 = df['total_bill'][:31]
no2 = df['tip'][51:78]

plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(3, 3, 1)
plt.plot(no1)
plt.title("Total Bill Line")

# Subplot 2
plt.subplot(3, 3, 3)
plt.plot(no2)
plt.title("Tip Line")

# Subplot 3
plt.subplot(3, 3, 5)
min_len = min(len(no1), len(no2))
plt.scatter(no1[:min_len], no2[:min_len])
plt.title("Scatter: Total Bill vs Tip")

# Subplot 4
plt.subplot(3, 3, 7)
plt.hist(no1)
plt.title("Total Bill Histogram")

# Subplot 5
plt.subplot(3, 3, 9)
plt.hist(no2)
plt.title("Tip Histogram")

plt.tight_layout()
plt.show()
```
