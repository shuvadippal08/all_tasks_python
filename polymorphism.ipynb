```python
from abc import ABC, abstractmethod
import logging



class Vehicle(ABC):
    def __init__(self):
        self.vehicle_id = input("Enter vehicle id: ")
        self.brand = input("Enter the brand: ")
        self._price_per_day = 0

    def set_price_per_day(self, price_per_day):
        self._price_per_day = price_per_day

    @abstractmethod
    def calculate_rent(self):
        pass


class Car(Vehicle):
    def __init__(self):
        super().__init__()
        self.set_price_per_day(1000)
        self.dayscount = int(input("Enter number of days: "))

    def calculate_rent(self):
        return self._price_per_day * self.dayscount


class Bike(Vehicle):
    def __init__(self):
        super().__init__()
        self.set_price_per_day(50)
        self.dayscount = int(input("Enter number of days: "))

    def calculate_rent(self):
        return self._price_per_day * self.dayscount


class Truck(Vehicle):
    def __init__(self):
        super().__init__()
        self.set_price_per_day(500)
        self.dayscount = int(input("Enter number of days: "))

    def calculate_rent(self):
        return self._price_per_day * self.dayscount
```

```python
car = Car()
print(car.calculate_rent())
```

```python
bike = Bike()
print(bike.calculate_rent())
```

```python
truck = Truck()
print(truck.calculate_rent())
```
