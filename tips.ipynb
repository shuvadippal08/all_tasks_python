```python
def membersize_dinner_tips(df):
    for size in sorted(df[df['time'] == 1]['size'].unique()):
        counts = df[
            (df['size'] == size) & (df['time'] == 1)
        ]['tips_range'].value_counts()

        print(f"\nMember size {size} visited in Dinner:")
        print(counts)
membersize_dinner_tips(df)
```

```python
def membersize_dinner_tips(df):
    for size in sorted(df[df['time'] == 1]['size'].unique()):
        counts = df[
            (df['size'] == size) & (df['time'] == 0)
        ]['tips_range'].value_counts()

        print(f"\nMember size {size} visited in Lunch:")
        print(counts)
membersize_dinner_tips(df)
```
