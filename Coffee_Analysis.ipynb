#### **Top 5 Countries based on production**

```python
k = {}
for country in df.Country.unique():
    k[country] = df[df.Country == country].Total_Production.max()
```

```python
for keys, values in k.items():
    j = sorted(k.values(),reverse=True)
    for i in range(0,5):
        if values == j[i]:
            print(f"{keys}: {values}")
```

#### Which Country has highest average export price

```python
k = {}
for country in df.Country.unique():
    k[country] = df[df.Country == country].Export_Price.mean()
    

for keys, values in k.items():
    if values == max(k.values()):
        print(f"{keys}: {values}") 
```

#### **Which country shows the highest growth in total production over the years?**

```python
yearly = df.groupby(['Country', 'Year'])['Total_Production'].sum().reset_index()
yearly
```

```python
growth = yearly.groupby('Country')['Total_Production'].agg(
    lambda x: x.iloc[-1] - x.iloc[0]
)
growth
```

```python
growth.idxmax(), growth.max()
```

#### **Relationship between each column**

```python
df[['Export_Price', 'Total_Production']].corr()
```

```python

plt.scatter(df['Export_Price'], df['Total_Production'])
plt.xlabel('Export Price')
plt.ylabel('Total Production')
plt.title('Export Price vs Total Production')
plt.show()
```

```python
import seaborn as sns
corr = df[['Export_Price', 'Total_Production']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Export Price and Total Production')
plt.xticks(rotation=90)
plt.yticks(rotation=60)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between all numeric features')
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.show()
```
