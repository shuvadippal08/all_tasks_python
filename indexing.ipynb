```python
a = input("Enter a character: ")
if a[0].isalpha():
    if a[0].isupper():
        print("Upper case")
    else:
        print("Lower case")
elif a[0].isdigit():
    print("Digit number")
else:
    print("Special Character")
```

```python
exp = int(input("Enter the number of years of experience: "))
sal = eval(input("Enter salary: "))
rating = eval(input("Enter the rating: "))
if exp >= 5:
    if rating >= 4:
        sal = sal * 12/10
        print("20'%' bonus")
    else:
        sal = sal * 11/10
        print("10'%' bonus")
else:
    print("No bonus")

print(sal)
```

```python
num = int(input("Enter a number: "))
if num >= 0:
    if num%2==0:
        print("Even number")
    else:
        print("Odd number")
else:
    print("Negative number")
```

```python
pin = int(input("Enter 4 digit the pin: "))
wam = int(input("Enter the amount for withdrwal: "))
bal = 52000
pink = 8945
if pin == pink:
    if bal >= wam:
        print("Withdrwal successfull")
    else:
        print("Enter amount lower / not available that much money")
else:
    print("PIN is incorrect")
```
