```python
from abc import ABC, abstractmethod


class Person(ABC):
    def __init__(self, name: str, age: int) -> None:
        self.__name = name
        self.__age = age

    def get_details(self):
        return f"Name: {self.__name}, Age: {self.__age}"

    @abstractmethod
    def role(self):
        pass


class Student(Person, ABC):
    def __init__(self, name: str, age: int, roll: int) -> None:
        super().__init__(name, age)
        self.__roll = roll

    def display(self):
        k = super().get_details()
        return f"{k}, Roll: {self.__roll}"

    def role(self):
        return "Student"

    @abstractmethod
    def is_sports(self) -> bool:
        pass

    @abstractmethod
    def is_academic(self) -> bool:
        pass


class Teacher(Person):
    def __init__(self, name: str, age: int, subject: str) -> None:
        super().__init__(name, age)
        self.subject = subject

    def role(self):
        return "Teacher"

    def display(self):
        return f"{self.get_details()}, Subject: {self.subject}"



class Academic:
    def __init__(self, subject: str, marks: float):
        self.subject = subject
        self.marks = marks

    def academic_details(self):
        return f"Subject: {self.subject}, Marks: {self.marks}"


class Sports:
    def __init__(self, sport_name: str, level: str):
        self.sport_name = sport_name
        self.level = level

    def sports_details(self):
        return f"Sport: {self.sport_name}, Level: {self.level}"


class SportsStudent(Student, Sports):
    def __init__(self, name, age, roll, sport_name, level):
        Student.__init__(self, name, age, roll)
        Sports.__init__(self, sport_name, level)

    def is_sports(self) -> bool:
        return True

    def is_academic(self) -> bool:
        return False

    def full_details(self):
        return f"{self.display()}\n{self.sports_details()}"

class AcademicStudent(Student, Academic):
    def __init__(self, name, age, roll, subject, marks):
        Student.__init__(self, name, age, roll)
        Academic.__init__(self, subject, marks)

    def is_sports(self) -> bool:
        return False

    def is_academic(self) -> bool:
        return True

    def full_details(self):
        return f"{self.display()}\n{self.sports_details()}"
    

class AllRounderStudent(Student, Academic, Sports):
    def __init__(self, name, age, roll,
                 subject, marks,
                 sport_name, level):

        Student.__init__(self, name, age, roll)
        Academic.__init__(self, subject, marks)
        Sports.__init__(self, sport_name, level)

    def is_sports(self) -> bool:
        return True

    def is_academic(self) -> bool:
        return True

    def full_details(self):
        return (
            f"{self.display()}\n"
            f"{self.academic_details()}\n"
            f"{self.sports_details()}"
        )


if __name__ == "__main__":

    teacher = Teacher("Mr. Roy", 45, "Math")
    print(teacher.display())

    print("\n-----------\n")

    sports_student = SportsStudent("Amit", 17, 101, "Football", "District")
    print(sports_student.full_details())

    print("\n-----------\n")

    all_rounder = AllRounderStudent(
        "Shuva", 18, 102,
        "Physics", 88,
        "Cricket", "State"
    )

    print(all_rounder.full_details())
```

```python
def move_zeros_to_end(arr):
    i=0
    for i in range(0,len(arr)):
        if arr[i]==0:
            for j in range(i+1,len(arr)):
                if arr[j] != 0:
                    arr[i],arr[j]= arr[j],arr[i]
                    break
    return arr
ls = [0,1,2,3,0,8]
print(move_zeros_to_end(ls))
```

```python
from abc import ABC, abstractmethod

class Person(ABC):
    def __init__(self, name: str):
        self.name = name          # Public
        self._role = "Person"     # Protected
        self.__secret = "ID123"   # Private

    @abstractmethod
    def get_role(self):
        pass

    def get_secret(self):
        return self.__secret
```

```python
class Student(Person):
    def __init__(self, name: str, marks: int):
        super().__init__(name)
        self.marks = marks

    def get_role(self):
        return "Student"
```

```python
class CollegeStudent(Student):
    def __init__(self, name: str, marks: int, branch: str):
        super().__init__(name, marks)
        self.branch = branch

    def get_details(self):
        return f"{self.name}, {self.branch}, {self.marks}"
```

```python
class Person:
    def __init__(self,name:str,age:int)->None:
        self.name = name
        self.age = age
# Heirarchical Inheritance

class Student(Person):
    def __init__(self,name:str,age:int, rollno:int)->None:
        super().__init__(name,age)
        self.rollno = rollno

# Multiple Inheritance
class Academic:
    def __init__(self,course:str,cgpa:float)->None:
        self.course = course
        self.cgpa = cgpa
class Sports:
    def __init__(self,sport_name:str,level:str)->None:
        self.sport_name = sport_name
        self.level = level

# Hybrid Inheritance

class AllrounderStudent(Student, Academic, Sports):
    def __init__(self, name:str,age:int,rollno:int,is_academic:bool = False,is_sport:bool=False,course:str=None, cgpa:float=None,sport_name:str=None, level:str=None):
        Student.__init__(self,name,age,rollno)
        self.is_academic = is_academic
        self.is_sport = is_sport

        if self.is_academic:
            Academic.__init__(self,course,cgpa)
        if self.is_sport:
            Sports.__init__(self,sport_name,level)
    
    def display(self):
        print(f"Name is {self.name}")
        print(f"Agae is {self.age}")
        print(f"Roll no is {self.rollno}")
        if self.is_academic:
            print(f"Academic Course is {self.course}")
            print(f"CGPA is {self.cgpa}")
        if self.is_sport:
            print(f"Sport name is {self.sport_name}")
            print(f"Level of sport is {self.level}")


```

```python
s1 = AllrounderStudent("Shuva",23,10300222167,is_academic=True,course="B.Tech",cgpa=9.09)
s1.display()
```

```python
class Person:
    def __init__(self):
        self.name = input("Enter name: ")
        self.age = int(input("Enter age: "))


class Student(Person):
    def __init__(self):
        super().__init__()
        self.sid = int(input("Enter student id: "))


class Academic:
    def __init__(self):
        self.course = input("Enter Academic course name: ")
        self.cgpa = float(input("Enter CGPA: "))


class Sports:
    def __init__(self):
        self.sport_name = input("Enter sport name: ")
        self.level = input("Enter level: ")


class AllRounderStudent(Student, Academic, Sports):
    def __init__(self, isAcademic=False, isSport=False):
        self.isAcademic = isAcademic
        self.isSport = isSport

        super().__init__()   # Student â†’ Person

        if self.isAcademic:
            Academic.__init__(self)

        if self.isSport:
            Sports.__init__(self)

    def display(self):
        print(f"Name = {self.name}, Age = {self.age}, ID = {self.sid}")

        if self.isAcademic:
            print(f"Academic: {self.course}, CGPA = {self.cgpa}")

        if self.isSport:
            print(f"Sport: {self.sport_name}, Level = {self.level}")


s1 = AllRounderStudent(isAcademic=True)
s1.display()
```
