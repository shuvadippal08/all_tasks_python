```python
import logging

logging.basicConfig(
    filename="logglearn.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s - %(name)s",
    filemode="a",
    force=True,
)

def add(a, b):
    logging.debug(f"Adding {a} and {b}")
    result = a + b
    logging.info(f"Result is {result}")
    return result

add(5, 3)


```
