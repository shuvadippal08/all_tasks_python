```python
class Foo:
    __a = int()
    __b = int()
    __c = int()
    def __init__(self):
        pass
    def setter(self,a,b,c):
        self.__a = a
        self.__b = b
        self.__c = c
    def getter(self):
        return self.__a, self.__b, self.__c
obj = Foo()
print(obj._Foo__a)
obj.setter(1,2,3)
print(obj._Foo__a)
print(obj.getter())
    
```
