```python
## PUBLIC MEMBERS

class Student:
    def __init__(self,name:str)->None:
        self.name = name
    
    def show(self):
        print(f"Name is {self.name}")

st = Student("Shuva")
print(st.name) # Accessible
st.show()
```

```python
## Protected
class Student:
    def __init__(self)->None:
        self._age = 20 # Protected Variable

class Child(Student):
    def display(self):
        print(self._age) # Accessible in sub class
par = Student()
print(par._age)
obj = Child()
obj.display()
print(obj._age)
```

```python
## Private

class Student:
    def __init__(self):
        self.__marks = 90
    def show(self):
        print(self.__marks)

st = Student()
# print(st.__marks) ## Cant be accessed Directly
st.show()
```

```python
##  Overriding
class Animal:
    def speak(self):
        print("Animal Speaks")

class Dog(Animal):
    def speak(self):
        print("Dog Barks")
a = Animal()
a.speak()

b = Dog()
b.speak()

```

```python
class Add:
    def add(self,a:int,b:int=None,c:int=None,d:int=None)->int:
        if b is None and c is None and d is None:
            return a
        elif c is None and d is None:
            return a+b
        elif d is None:
            return a+b+c
        else:
            return a+b+c+d

a = Add()
print(a.add(1))
print(a.add(1,2))
print(a.add(1,2,3))
print(a.add(1,2,3,4))
```

```python
class Parent:
    def __init__(self):
        self.__salary = 50000   # Private variable

    def show_salary(self):
        print("Salary:", self.__salary)


class Child(Parent):
    def display(self):
        print("Salary from child:", self._Parent__salary)

a = Parent()
a.show_salary()
# print(a.__salary) ## this will show an error

c = Child()
c.display()
```
