## **Start with matplot lib**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1,50,10)
y = np.random.randint(1,50,10)
print(x,y,sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='chartreuse',linestyle='-.')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5,marker='v')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='red')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='cyan',markersize='10')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='cyan',markersize='10',markeredgecolor='yellow')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=1,marker='*',markerfacecolor='cyan',markersize='10',markeredgecolor='yellow',linewidth=3)
plt.show()

```

```python
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)
print(x,y,sep='\n')
```

```python
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()

plt.subplot(1, 2, 2)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='cyan')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.grid(True)

plt.subplot(3, 3, 1)
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')
plt.grid(True)



plt.subplot(3, 3, 5)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='teal')
plt.plot(x,z,label='cosq',color='cyan')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')
plt.grid(True)


plt.subplot(3, 3, 9)
plt.title("SINQ and COSQ3 graph")
plt.plot(x,y,label='sinq',color='red')
plt.plot(x,z,label='cosq',color='green')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')

plt.grid(True)
plt.subplots_adjust(hspace=0.8,wspace=0.8)
plt.tight_layout()
plt.show()
```
