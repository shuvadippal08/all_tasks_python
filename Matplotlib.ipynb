## **Start with matplot lib**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1,50,10)
y = np.random.randint(1,50,10)
print(x,y,sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='r',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='chartreuse',linestyle='-.')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5,marker='v')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='cyan',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='red')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='cyan',markersize='10')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=0.5,marker='*',markerfacecolor='cyan',markersize='10',markeredgecolor='yellow')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("random x")
plt.ylabel("random y")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.',alpha=1,marker='*',markerfacecolor='cyan',markersize='10',markeredgecolor='yellow',linewidth=3)
plt.show()

```

```python
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)
print(x,y,sep='\n')
```

```python
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()

plt.subplot(1, 2, 2)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='cyan')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.grid(True)

plt.subplot(3, 3, 1)
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')
plt.grid(True)



plt.subplot(3, 3, 5)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='teal')
plt.plot(x,z,label='cosq',color='cyan')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')
plt.grid(True)


plt.subplot(3, 3, 9)
plt.title("SINQ and COSQ3 graph")
plt.plot(x,y,label='sinq',color='red')
plt.plot(x,z,label='cosq',color='green')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.legend(loc='upper right')

plt.grid(True)
plt.subplots_adjust(hspace=0.8,wspace=0.8)
plt.tight_layout()
plt.show()
```

## **Scatter plot**

```python
x = np.arange(1,11)
y = np.random.randint(1,100,10)
print(x,y,sep='\n')
```

```python
plt.scatter(x,y)
plt.show()
```

```python
company = ['a','b','c','d']
revenew = [11,34,45,54]
```

```python
plt.scatter(company,revenew)

## to find the correlation between plots (numerical and analytical data)
```

```python
plt.bar(company,revenew)
```

```python
company = ['a','b','c','d']
revenew = [11,34,45,54]
plt.bar(company,revenew,color=['blue','orange','r','y'])
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("revenew")
```

```python
company = ['a','b','c','d']
revenew = [11,34,45,54]
plt.barh(company,revenew,color=['blue','orange','r','y'])
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("revenew")
plt.show()
```

## **Histogram**

```python
company = [12,15,20,22]
revenew = [11,34,45,54]
plt.hist(company)
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("revenew")
plt.show()
```

## Pie plot

```python
plt.pie(revenew,labels=company,autopct="%.1f%%")
plt.title("Pie plot")
plt.show()
```

```python
company = ['A','B',"C","D","E","F"]
revenew = [11,54,78,45,30,89]
plt.plot(company,revenew)
plt.title("Line Plot")
plt.ylabel("revenew")
plt.xlabel("Company")
plt.show()
```

## Box plot for outliers

```python
plt.boxplot(revenew)
plt.title("Box plot")
plt.show()
```
